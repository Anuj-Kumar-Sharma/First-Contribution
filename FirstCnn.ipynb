# Building CNN - Convolutional Neural Network

### Load Dataset

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import tensorflow.keras as tf


````

````python
# change dimension
# check dimension
X_train.ndim
# this data is in 3D but CNN wants 4D
````

````python
# Change dimension
X_train = np.expand_dims(X_train, -1)
X_test = np.expand_dims(X_test, -1)
````

````python
# check dim
X_train.ndim
````

## feature scaling

````python
# feature scaling that term i don't know
X_train = X_train/255
X_test = X_test/255
````

````python
# after scaling
X_train[0]
````

````python
# traing ke sath sath testing bhi karna hoga thats why we want validation data type
````

## split dataset

````python
# split dataset
from sklearn.model_selection import train_test_split 
X_train, X_validation, y_train, y_validation = train_test_split(X_train, y_train, test_size= 0.2, random_state=0)
# test_size is how many % of data you want to give in validation datatype and, random_state is random data and size 0.2 is 20% data we want in validation.
````

````python
# sahpe
X_train.shape,  y_train.shape #, X_validation.shape, y_validation.shape
````

````python
# show validation data set
X_validation.shape, y_validation.shape
````

#  Convolution Neural Network Model Building

````python
model =  keras.models.Sequential([
                         keras.layers.Conv2D(filters=32, kernel_size=3, strides=(1, 1), padding='valid', activation='relu', input_shape=(28,28,1)),
                         keras.layers.MaxPool2D(pool_size=(2,2)),
                         keras.layers.Flatten(),
                         #creat neural network
                         #craeat input layers
                         keras.layers.Dense(units=128, activation='relu'),
                        #  units is how many neuron you want
                        #  creal output layers
                         keras.layers.Dense(units=10, activation='softmax'), # activation=softmax gives probability

                         
])
````

````python
# if you want summary of your model
model.summary()
````

````python
# compile
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
````

````python
# train model
model.fit(X_train, y_train, epochs=30, batch_size=512, verbose=1, validation_data=(X_validation,y_validation))
# fit is method to train model
````

# Test model

````python
# model.predict(X_test[0])
# here we can face error because our test data is in 3D but they want in 4D.
model.predict(np.expand_dims(X_test[3], axis=0)).round(2)
````

````python
# for single max value 
np.argmax(model.predict(np.expand_dims(X_test[3], axis=0)).round(2))
````

````python
# if you want to crosscheck
y_test[3]
````

````python
y_pred =  model.predict(X_test).round(2)
y_pred
````

````python
# check accuracy
model.evaluate(X_test, y_test)
````

````python
plt.figure(figsize=(16,30))
 
j=1
for i in np.random.randint(0, 1001,60):
  plt.subplot(10,6, j); j+=1
  plt.imshow(X_test[i].reshape(28,28), cmap = 'Greys')
  plt.title('Actual = {} / {} \nPredicted = {} / {}'.format(class_labels[y_test[i]], y_test[i], class_labels[np.argmax(y_pred[i])],np.argmax(y_pred[i])))
  plt.axis('off')
````

````python
plt.figure(figsize=(16,16))
 
j=1
for i in np.random.randint(0, 1000,25):
  plt.subplot(5,5, j); j+=1
  plt.imshow(X_test[i].reshape(28,28), cmap = 'Greys')
  plt.title('Actual = {} / {} \nPredicted = {} / {}'.format(class_labels[y_test[i]], y_test[i], class_labels[np.argmax(y_pred[i])],np.argmax(y_pred[i])))
  plt.axis('off')
````

# Confusion matrix


````python
from sklearn.metrics import confusion_matrix
 
plt.figure(figsize=(16,9))
y_pred_labels = [ np.argmax(label) for label in y_pred ]
cm = confusion_matrix(y_test, y_pred_labels)
# show cm 
sns.heatmap(cm, annot=True, fmt='d',xticklabels=class_labels, yticklabels=class_labels)
````

### precision =  true positive / (true positive + false positive)
### recall   =  true positive / (true positive  + false negative)

````python
from sklearn.metrics import classification_report
cr= classification_report(y_test, y_pred_labels, target_names=class_labels)
print(cr)
````
